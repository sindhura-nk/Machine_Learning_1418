Unsupervised Learning : Principal Component Analysis

```python
path=r'https://raw.githubusercontent.com/sindhura-nk/Datasets/refs/heads/main/laptopPrice.csv'
import pandas as pd
df = pd.read_csv(path)
df.head()
```

```python
X = df.drop(columns='Price')
Y = df[['Price']]
```

```python
X.head()
```

```python
Y.head()
```

## Data cleaning and Data preprocessing

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
cat_pipe = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(handle_unknown='ignore',sparse_output=False)
)
con_pipe = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler()
)
pre = ColumnTransformer([
    ('cat',cat_pipe,cat),
    ('con',con_pipe,con)
]).set_output(transform='pandas')
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
X_pre_pca = pca.fit_transform(X_pre)
X_pre_pca
```

```python
pca.components_
```

```python
pca2 = PCA(n_components=3)
X_pre_pca2 = pca.fit_transform(X_pre)
X_pre_pca2
```

## train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_pre_pca,Y,test_size=0.2,random_state=1)
```

```python
X.shape
```

```python
xtrain.shape
```

```python
xtrain2,xtest2,ytrain2,ytest2 = train_test_split(X_pre_pca2,Y,test_size=0.2,random_state=1)
```

```python
xtrain2.shape
```

## Model building

```python
from sklearn.ensemble import RandomForestRegressor
rg = RandomForestRegressor(max_depth=10,min_samples_leaf=3,min_samples_split=4,n_estimators=15)
rg.fit(xtrain,ytrain)
```

```python
rg.score(xtrain,ytrain)
```

```python
rg.score(xtest,ytest)
```
