## Build the Apriori Model
### Criteria: Set the minimum values for Confidence, Lift and Support
    Support: 7*5/(len(df))
    Confidence:0.2
    Lift:3

Bi-weekly data, transaction per day are 10
support: 14*10/len(df)

```python
# generator function example
range(2,10)
```

```python
list(range(2,10))
```

```python
tuple(range(2,10))
```

```python
from apyori import apriori
sup = 7*5/(len(df))
conf = 0.2
lift = 3
association_rules = apriori(records,min_support=sup,min_confidence=conf,min_lift=lift) # this is a generator function.
association_rules = list(association_rules) # type cast association rules so that all the values can be captured
association_rules[0]
```

```python
for item in association_rules:
  #print(item)
  pair = item[0]
  items = [i for i in pair]
  print(f'Associated Items: {items[0]} ==> {items[1]}')

  sup = str(item[1])
  print(f'Support : {sup}')

  conf = str(item[2][0][2])
  lift = str(item[2][0][3])

  print(f'Confidence : {conf}')
  print(f'Lift : {lift}')

  print('===============================')

```

```python
results =[]
for item in association_rules:
  #print(item)
  pair = item[0]
  items = [i for i in pair]
  print(f'Associated Items: {items[0]} ==> {items[1]}')

  sup = str(item[1])
  print(f'Support : {sup}')

  conf = str(item[2][0][2])
  lift = str(item[2][0][3])

  print(f'Confidence : {conf}')
  print(f'Lift : {lift}')

  print('===============================')
  results.append([items[0],items[1],sup,conf,lift])
print(results)

```

```python
results
```

```python
df_associations = pd.DataFrame(results,columns=['Product1','Product2','Support','Confidence','Lift'])
df_associations
```

```python
df_associations.to_csv('Product Associations.csv',index=False)
```
